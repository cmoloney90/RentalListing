```python
import json
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from ggplot import *

#read in train as test data
with open('train.json') as json_data:
    train_json = json.load(json_data)
    
train_df = pd.DataFrame(train_json)

with open('test.json') as json_data:
    test_json = json.load(json_data)
    
test_df = pd.DataFrame(test_json)

print("Train Row Count: ", train_df.shape[0])
print("Test Row Count: ", test_df.shape[0])
train_df.columns
```

```python
import seaborn as sns

#check out target variable
il = ggplot(train_df, aes(x='interest_level')) +\
    geom_bar()
il.show()
```

Interest level is low for the majority of cases.

Next I'm going to look at the other columns - Bedrooms, Bathrooms, Price and coordinates

```python
#Bedrooms
bed_bar = ggplot(train_df,aes(x='bedrooms', fill='interest_level')) +\
        geom_bar()
    
bed_violin = ggplot(train_df,aes(x='bedrooms', color='interest_level')) +\
        geom_density()

bed_bar.show()
bed_violin.show()
```

```python
#Bathrooms
bath_bar = ggplot(train_df,aes(x='bathrooms', fill='interest_level')) +\
            geom_bar()
bath_bar.show()
```

```python
upper = np.percentile(train_df.latitude.values, 99)
lower = np.percentile(train_df.latitude.values, 1)
train_df['latitude'].ix[train_df['latitude']<lower] = lower
train_df['latitude'].ix[train_df['latitude']>upper] = upper

upper = np.percentile(train_df.longitude.values, 99)
lower = np.percentile(train_df.longitude.values, 1)
train_df['longitude'].ix[train_df['longitude']<lower] = lower
train_df['longitude'].ix[train_df['longitude']>upper] = upper

g = ggplot(train_df, aes(x='latitude', y='longitude', color='interest_level')) +\
        geom_point(alpha = 0.5)
g.show()
```
