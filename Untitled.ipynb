```python
upper = np.percentile(train_df.latitude.values, 99)
lower = np.percentile(train_df.latitude.values, 1)

train_df.loc[train_df['latitude']<lower, 'latitude'] = lower
train_df.loc[train_df['latitude']>upper, 'latitude'] = upper

upper = np.percentile(train_df.longitude.values, 99)
lower = np.percentile(train_df.longitude.values, 1)

train_df.loc[train_df['longitude']<lower, 'longitude'] = lower
train_df.loc[train_df['longitude']>upper, 'longitude'] = upper

g = ggplot(train_df, aes(x='latitude', y='longitude', color='interest_level')) +\
        geom_point(alpha = 0.5)
g.show()
```
